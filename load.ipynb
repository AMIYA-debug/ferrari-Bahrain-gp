```python
import fastf1
import numpy as np
import pandas as pd

fastf1.Cache.enable_cache('cache')
session=fastf1.get_session(2024,'Bahrain','R')
session.load()
```

```python
df=pd.read_csv('leclerc_2024_Bahrain_R_expanded.csv')
df.head()
```

```python
df.drop(columns=['pit_duration_s','pit_stop_dur_s','avg_tyre_temp','lap_number','gap_to_leader_s','lap_time_raw','s1_raw','s2_raw','s3_raw'],inplace=True)
```

```python
df.dropna(axis=0,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
session_time_seconds=[]
for time in df['session_time_raw']:
   time=pd.to_timedelta(time).total_seconds()
   session_time_seconds.append(time)

df['session_time_sec']=session_time_seconds  

   
```

```python
df.drop(columns=['session_time_raw'],inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
df['compound']=le1.fit_transform(df['compound'])
le2=LabelEncoder()
df['pit_in']=le2.fit_transform(df['pit_in'])
le3=LabelEncoder()
df['pit_out']=le3.fit_transform(df['pit_out'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
x=df.drop(columns=['compound'])
y=df['compound']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
col_to_scale=['lap_time_s','s1_s','s2_s','s3_s','tyre_life','position','prev_position','avg_speed','max_speed','std_speed','avg_throttle','avg_brake','session_time_sec','stint','overtakes']
x_train[col_to_scale]=scaler.fit_transform(x_train[col_to_scale])
x_train
```

```python
x_test[col_to_scale]=scaler.transform(x_test[col_to_scale])
```

```python
import pickle
pickle.dump(scaler,open('scaler.pkl','wb'))
```

```python
pickle.dump(le1,open('compound.pkl','wb'))
pickle.dump(le2,open('pit_in.pkl','wb'))
pickle.dump(le3,open('pit_out.pkl','wb'))
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping , TensorBoard
import datetime
```

```python
model=Sequential(
  [
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(64,activation='relu'),
    Dense(16,activation='relu'),
    Dense(1,activation='sigmoid') # if multiclass use softmax
  ]
)
```

```python
model.summary()
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.001)
loss=tf.keras.losses.binary_crossentropy
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%M%d-%H%M%S")
TensorBoard_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
EarlyStopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
history=model.fit(
  x_train,y_train,
  validation_split=0.2,
  epochs=100,
  batch_size=32,
  verbose=1,
  callbacks=[EarlyStopping_callback,TensorBoard_callback]

)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
for ele in y_pred:
  if ele >=0.5 :
    print('soft')
  else:
    print('hard')  
```

```python
def build_model(hp):
    model = tf.keras.Sequential()

    for i in range(hp.Int('num_layers', min_value=1, max_value=4)):
        
        units_count = hp.Int(f'units_{i}', min_value=32, max_value=128, step=32)
        model.add(tf.keras.layers.Dense(units=units_count, activation='relu'))
        
        rate=hp.Float(f'dropout_{i}', min_value=0.2, max_value=0.5, step=0.1)
        model.add(tf.keras.layers.Dropout(rate=rate))

    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
    
    hp_learning_rate = hp.Choice('learning_rate', values=[0.01, 0.001, 0.0001])
    
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp_learning_rate),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    return model
```

```python

import keras_tuner as kt
early=tf.keras.callbacks.EarlyStopping(patience=6,restore_best_weights=True)
tuner = kt.Hyperband(
    build_model,
    objective='val_accuracy',
    max_epochs=50,
    factor=3,
    directory='log_dir',
    project_name='ferrari3'
)
tuner.search(
    x_train, y_train,
    epochs=50,
    validation_split=0.15,
    callbacks=[early]
)
```

```python
best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
best_model=tuner.hypermodel.build(best_hps)
```

```python
best_model.fit(
   x_train,y_train,
   validation_split=0.2,
   epochs=20,
   callbacks=[early]
)
```

```python
y_pred=best_model.predict(x_test)
for ele in y_pred:
  if ele >=0.5 :
    print('soft')
  else:
    print('hard') 
```
